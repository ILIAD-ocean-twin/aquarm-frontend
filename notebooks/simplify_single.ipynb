```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('AIGBR36-trajectories-opendrift_passive_drift_mpa_uk_2024_2025-10-09T140604.json')
```

```python
gdf = gdf.iloc[::4,:].reset_index(drop=True)
```

```python
gdf['geometry'] = gdf.geometry.simplify(tolerance=0.1)
```

```python
gdf.to_file('AIGBR36.geojson', driver='GeoJSON')
```
