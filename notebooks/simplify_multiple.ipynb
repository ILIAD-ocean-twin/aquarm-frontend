```python
import geopandas as gpd
import glob
from tqdm.notebook import tqdm
```

```python
fnames = glob.glob("trajectories_in/*.json")
fnames = sorted(fnames)
```

```python
# for fname in tqdm(fnames[:8]):
for fname in tqdm(fnames):
    # print(site_id)
    site_id = fname.split('/')[1].split('-')[0]
    gdf = gpd.read_file(fname)
    gdf = gdf.iloc[::24,:].reset_index(drop=True)
    gdf['geometry'] = gdf.geometry.simplify(tolerance=0.001)
    gdf.to_file(f"trajectories_out/{site_id}.geojson", driver='GeoJSON')
```
